```python
# importing matplotlib module
from matplotlib import pyplot as plt
 
# Y-axis values
y = [10, 5, 8, 4, 2]
 
# Function to plot
# histogram (blue)
plt.hist(y)
# line graph
#if one array is given it is taken as y-axis with x-axis as defaults([0,1,2,3...]),(orange)
plt.plot(y)  
# both x and y axis can also be defined (green)
plt.plot([1,2,3,4,5],[1,4,9,16,25])  

plt.title('Graph')  
plt.ylabel('Y axis')  
plt.xlabel('X axis')

# Function to show the plot  
plt.show()  
```

```python
# formating style 

from matplotlib import pyplot as plt  

plt.plot([2, 2, 3, 4, 5], [1, 3, 8, 8 ,2], '-b')  #default-blue solid line

plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], 'ro')  #red dots

plt.plot([2, 3, 4, 5], [3, 9, 9, 17], '-g')  #green solid line 

plt.plot([1, 2, 3, 4], [7, 12, 13, 17.5], '--')  #dashed line with default color

plt.plot([1, 2, 3, 5], [12, 8, 1, 7], '^k:')  # Black triangle up markers connected by a dotted line

# g,r,c,m,y,k,w any color abbreviation can be used

plt.axis([0, 6, 0, 20])  #x and y axis scaling
plt.show()  
```
